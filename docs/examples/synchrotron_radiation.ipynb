# Synchrotron Radiation

Minimal demonstration of the synchrotron radiation functions.

```python
import numpy as np
import matplotlib.pyplot as plt

from pmd_beamphysics.radiation.synchrotron_radiation import (
    S_fast,
    S_exact,
    S_benchmarking,
)
```

```python
S_exact(1)
```

```python
# Single value
S_fast(1)
```

```python
# Array input
xi = 2 ** np.linspace(-10, 3, 100)
s0 = S_exact(xi)
s1 = S_fast(xi)
rel_err = s1 / s0 - 1
```

```python
%%timeit
S_exact(xi)
```

```python
%%timeit
S_fast(xi)
```

```python
fig, ax = plt.subplots()
ax.plot(xi, s0, label="exact")
ax.plot(xi, s1, "--", label="fast")
ax.legend()

ax2 = ax.twinx()
ax2.plot(xi, rel_err)

ax.set_xlabel(r"$\xi$")
ax.set_ylabel(r"$S(\xi)$")

ax2.set_yscale("log")
ax.set_xscale("log")
ax.set_yscale("log")
```
